```python
import pandas as pd
import requests

# API 키 입력
API_KEY = 'AIzaSyAK4U944sucnjBmyEixJRIxijDt1ozcqhc'

# 번역 함수
def translate_text(text, target_lang='en'):
    if pd.isna(text):  # NaN 값 처리
        return text

    url = "https://translation.googleapis.com/language/translate/v2"

    params = {
        'q': text,
        'target': target_lang,
        'key': API_KEY
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        result = response.json()
        return result['data']['translations'][0]['translatedText']
    else:
        print(f"Error: {response.status_code}")
        return text  # 에러 발생 시 원본 텍스트 반환

# CSV 파일 번역 함수 (앞 5개 행만 번역)
def translate_csv_limited(input_csv, target_lang='en'):
    df = pd.read_csv(input_csv)

    # factor 열만 번역 수행, 앞 5개 행에 대해서만 번역
    df['en_factor'] = df['factor'].iloc[:limit].apply(lambda x: translate_text(str(x), target_lang))

    # 번역된 결과 확인 (5개 행만 표시)
    return df.head(limit)

# 파일을 수동으로 경로 지정해서 사용
input_csv = './Final_Flexible_Matching_DataFrame_with_Correct_en_factor.csv'  # 파일 경로 입력
translated_df = translate_csv_limited(input_csv)

# 번역된 결과 출력
print(translated_df)
```
